```python
import numpy as np
from keras.models import model_from_json
from keras.optimizers import SGD
from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator
#from sklearn.metrics import roc_auc_score

#from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.externals import joblib
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
import scikitplot as skplt
import h5py

import numpy as np
import os
import cv2
#import scikitplot as skplt
import tensorflow as tf

from keras.models import Sequential, Model
from keras.layers import Conv2D, ZeroPadding2D, MaxPooling2D, GlobalMaxPooling2D, Input
from keras.layers.core import Flatten, Dense, Dropout, Lambda
from keras import applications

from sklearn.metrics import roc_auc_score,roc_curve
from sklearn.metrics import roc_curve
#import scikitplot as skplt
import matplotlib.pyplot as plt

###Keras
import keras
from keras.utils import np_utils
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input
from keras.layers import Dense, Activation, Flatten
from keras.layers import merge, Input
from keras.layers import MaxPooling2D
from keras.models import Model
from sklearn.utils import shuffle
#from sklearn.cross_validation import train_test_split
from keras.optimizers import SGD
from keras.preprocessing.image import ImageDataGenerator

###Sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.externals import joblib
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.utils import shuffle
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.externals import joblib

import pandas as pd
import os
import numpy as np
from sklearn import metrics
from scipy.stats import zscore
from sklearn.model_selection import KFold
from keras.layers import GlobalMaxPooling2D

from keras.utils import np_utils
from keras.callbacks import Callback, EarlyStopping
from keras.callbacks import ModelCheckpoint
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
```

```python
PATH=os.getcwd()
#Giving the path of the directory
data_path='C:/Users/tusha/Documents/MasterThesis/Classification/data/dm'
data_dir_list=os.listdir(data_path)##getting the list of the folder in the directory.
print(data_dir_list)

img_data_list=[] #making list to store all the images

for dataset in data_dir_list:
    img_list=os.listdir(data_path+'/'+dataset) #taking the list of the images in each folder
    print('Loaded the image of dataset-'+ '{}\n'.format(dataset))
    
    for img in img_list:
        img_path=data_path+'/'+dataset+'/'+img #getting the image path
        img2 = cv2.imread(img_path)
        img2 = cv2.resize(img2,(224,224))
        img2 = np.tile(img2,(1,1,1))
        #print('input image shape:',img2.shape)
        #cv2.imwrite("original2.png",img2)
        x=image.img_to_array(img2) #converting the image to array
        x=np.expand_dims(x,axis=0) #putting them in the row axis
        x=preprocess_input(x) #preprocessing
        img_data_list.append(x)

img_data = np.array(img_data_list)
#img_data = img_data.astype('float32')
print (img_data.shape)
img_data=np.rollaxis(img_data,1,0)
print (img_data.shape)
img_data=img_data[0]
print (img_data.shape)


num_classes=2
print('Number of classes=',num_classes)
num_of_samples=img_data.shape[0]
print('Number of samples=',num_of_samples)
labels=np.ones((num_of_samples,),dtype='int64')


labels[0:1786]=1
labels[1786:3108]=0


#Suufle the data 
x,y=shuffle(img_data,labels,random_state=2)

```

```python
from sklearn.model_selection import train_test_split
#split the training and test set
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.20,random_state=2)
print('Training data:',x_train.shape,'Validatuin data=',x_val.shape)
print('Training labels:',y_train.shape,'Validation data=',y_val.shape)

from keras.layers import MaxPooling2D

y_val1=np_utils.to_categorical(y_val,num_classes)
y_train1=np_utils.to_categorical(y_train,num_classes)

###Defining_AUC for each Epoch
def auc_roc(y_true, y_pred):
    # any tensorflow metric
    value, update_op = tf.contrib.metrics.streaming_auc(y_pred, y_true)

    # find all variables created for this metric
    metric_vars = [i for i in tf.local_variables() if 'auc_roc' in i.name.split('/')[1]]

    # Add metric variables to GLOBAL_VARIABLES collection.
    # They will be initialized for new session.
    for v in metric_vars:
        tf.add_to_collection(tf.GraphKeys.GLOBAL_VARIABLES, v)

    # force to update metric values
    with tf.control_dependencies([update_op]):
        value = tf.identity(value)
        return value

##InputShape
image_input=Input(shape=(224,224,3))

###Casll the Model
Resnet50model=keras.applications.resnet50.ResNet50(input_tensor=image_input,include_top=False,weights='imagenet')
Resnet50model.summary()

from keras.layers import AveragePooling2D


from keras.metrics import categorical_accuracy


last_layer = Resnet50model.output
z= AveragePooling2D(pool_size=(7, 7))(last_layer)
z= keras.layers.GlobalMaxPooling2D()(z)
out = Dense(num_classes, activation='softmax', name='output_layer')(z)
custom_model = Model(inputs=image_input,outputs= out)

custom_model.summary()

##Initializing weights and models name to save
Inceptionv3_model_name ="Resnet2_keras_model.json"
Inceptionv3_normalizer_name = "Resnet2_normalizer.pkl"
filepath= 'Resnet2_weight.hdf5'

datagen = ImageDataGenerator(
    featurewise_center=False,
    featurewise_std_normalization=False,
    rotation_range=0,
    vertical_flip=False)
datagen.fit(x_train)



batch_size=32
###Check-point
filepath= 'Resnet2_weight.hdf5'
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint,EarlyStopping(monitor='auc_roc', patience=300, verbose=1, mode='max')]

#### fixed lr part
gradientDescent = SGD(lr=0.01, decay=0.96)
custom_model.compile(gradientDescent, loss= 'binary_crossentropy', metrics = ['categorical_accuracy', auc_roc])
custom_model.fit_generator(datagen.flow(x_train, y_train1, batch_size=batch_size),steps_per_epoch=98,
                               validation_data=(x_val, y_val1),
                               epochs=2, verbose=1, workers=4,callbacks=callbacks_list)

############################### #serialize model to JSON and weights to HDF5
model_json = custom_model.to_json()
with open(Inceptionv3_model_name, 'w') as json_file:
     json_file.write(model_json)
joblib.dump(datagen,Inceptionv3_normalizer_name)
print("model and weights have been saved")

###############################
```

```python
def plot_roc(pred,y):
    fpr, tpr, _ = roc_curve(y, pred)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc="lower right")
    plt.show()
```
