# Visualization

```python
def get_model(net):

    if net=='resnet50':
        json_file = open('Model-Fold-1.json', 'r')
        loaded_json_model = json_file.read()
        json_file.close()

        loaded_model = model_from_json(loaded_json_model)
        loaded_model.load_weights('weights-best_model-Fold-1.hdf5')
        print("model loaded")
        loaded_model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
    return loaded_model
    
```

```python
img2 = cv2.imread('1742_3_0236_.png')
img2 = cv2.resize(img2,(224,224))
img2 = np.tile(img2,(1,1,1))
x=image.img_to_array(img2) #converting the image to array
x=np.expand_dims(x,axis=0) #putting them in the row axis
x=preprocess_input(x)
```

```python
from keras.models import Model

def auc_roc(y_true, y_pred):
    # any tensorflow metric
    value, update_op = tf.contrib.metrics.streaming_auc(y_pred, y_true)

    # find all variables created for this metric
    metric_vars = [i for i in tf.local_variables() if 'auc_roc' in i.name.split('/')[1]]

    # Add metric variables to GLOBAL_VARIABLES collection.
    # They will be initialized for new session.
    for v in metric_vars:
        tf.add_to_collection(tf.GraphKeys.GLOBAL_VARIABLES, v)

    # force to update metric values
    with tf.control_dependencies([update_op]):
        value = tf.identity(value)
        return value
```

```python
json_file = open('Model-Fold-1.json', 'r')
loaded_json_model = json_file.read()
json_file.close()

loaded_model = model_from_json(loaded_json_model)
loaded_model.load_weights('weights-best_model-Fold-1.hdf5')

print("model loaded")
loaded_model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

####### Loading NORMALIZER #######
########                 ################################
normalizer  = joblib.load('Normalizer-Fold-1.pkl')
normalized_Xtest = normalizer.standardize(x)
loaded_model.summary()
```

```python
model=get_model('resnet50')
```

```python
layer_outputs = [layer.output for layer in loaded_model.layers] 
layer_outputs = layer_outputs[1:] 
activation_model = Model(inputs=loaded_model.input, outputs=layer_outputs)
activations = activation_model.predict(normalized_Xtest.reshape(1,224,224,3))
 
def display_activation(activations, col_size, row_size, act_index): 
    activation = activations[act_index]
    activation_index=0
    fig, ax = plt.subplots(row_size, col_size, figsize=(row_size*2.5,col_size*1.5))
    for row in range(0,row_size):
        for col in range(0,col_size):
            ax[row][col].imshow(activation[0, :, :, activation_index], cmap='gray')
            activation_index += 1
```
